Fundamentos de Rust: Seguridad y Rendimiento

Bienvenido a esta guía interactiva de Rust. Rust es un lenguaje de programación que se centra en tres objetivos: seguridad, velocidad y concurrencia.

A diferencia de otros lenguajes, Rust no utiliza un recolector de basura (Garbage Collector). En su lugar, garantiza la seguridad de la memoria en tiempo de compilación mediante un sistema de reglas que el compilador verifica estrictamente. Si tu código compila, es muy probable que sea seguro.


1. Variables y Mutabilidad

En Rust, las variables son inmutables por defecto. Esto significa que una vez que asignas un valor a un nombre, ese valor no puede cambiar. Esto promueve la seguridad al evitar cambios accidentales en el estado de tu programa.

    • let: Declara una variable inmutable.
    • let mut: Declara una variable mutable.
    • Shadowing (Sombreado): Rust permite declarar una nueva variable con el mismo nombre que una anterior, "sombreándola". Esto es útil para transformar datos sin necesidad de crear nombres como data_string y data_int.


```rust
// 1. Inmutabilidad
let x = 5;
// x = 6; // Esto daría un error de compilación

// 2. Mutabilidad
let mut y = 10;
println!("Valor original de y: {}", y);
y = 15; // Esto es válido
println!("Nuevo valor de y: {}", y);

// 3. Shadowing
let espacios = "   "; // Tipo string
let espacios = espacios.len(); // Ahora es tipo numérico (usize)
println!("Cantidad de espacios: {}", espacios);
```

2. Tipos de Datos

Rust es un lenguaje estáticamente tipado, lo que significa que debe conocer los tipos de todas las variables en tiempo de compilación.

Tipos Escalares
Representan un valor único:

    • Enteros: i32 (por defecto), u32 (sin signo), i64, etc.
    • Flotantes: f64 (precisión doble).
    • Booleanos: true y false.
    • Caracteres: char (representa un valor Unicode de 4 bytes).

Tipos Compuestos
Agrupan múltiples valores:

    • Tuplas: Tienen una longitud fija y pueden contener tipos diferentes.
    • Arrays (Arreglos): Tienen una longitud fija y todos los elementos deben ser del mismo tipo.

```rust
// Escalares
let entero: i32 = -42;
let flotante: f64 = 3.1416;
let es_rust_genial: bool = true;
let emoji: char = '🦀';

// Compuestos
let tupla: (i32, f64, char) = (500, 6.4, 'R');
let (a, b, c) = tupla; // Destructuración

let array: [i32; 3] = [10, 20, 30];
println!("Primer elemento del array: {}", array[0]);
```

3. Funciones

Las funciones se declaran con la palabra clave fn. Rust utiliza la convención snake_case para los nombres.

Nota Clave: En Rust, existe una diferencia entre sentencias (que terminan en ; y no devuelven valor) y expresiones (que evalúan un valor y no llevan ;). La última línea de una función suele ser una expresión que se devuelve implícitamente.

```rust
fn suma(a: i32, b: i32) -> i32 {
    a + b // Expresión: nota que no tiene punto y coma al final
}

let resultado = suma(10, 5);
println!("El resultado de la suma es: {}", resultado);
```

4. Control de Flujo

El flujo de ejecución en Rust es muy similar al de otros lenguajes, pero con algunas mejoras de seguridad.

    • if: Las condiciones deben ser booleanas (Rust no tiene el concepto de "truthy" o "falsy" como JS o Python).
    • loop: Un bucle infinito que se rompe con break.
    • while: Se ejecuta mientras una condición sea verdadera.
    • for: Ideal para iterar sobre colecciones de forma segura.

```rust
let numero = 7;

// If / Else
if numero < 10 {
    println!("Es un número pequeño");
} else {
    println!("Es un número grande");
}

// Bucle for sobre un rango (del 1 al 3)
for n in 1..4 {
    println!("Iteración número: {}", n);
}
```

5. El Concepto Maestro: Ownership (Propiedad)

Este es el concepto más innovador de Rust. El Ownership permite a Rust gestionar la memoria sin un recolector de basura.

Reglas básicas:
    1. Cada valor en Rust tiene una variable que es su dueño (owner).
    2. Solo puede haber un dueño a la vez.
    3. Cuando el dueño sale del ámbito (scope), el valor se descarta automáticamente.

Cuando pasas una variable a una función o la asignas a otra, el valor se mueve (Move), a menos que el tipo implemente una copia profunda o uses referencias (&).

```rust
let s1 = String::from("Hola");
let s2 = s1; // El valor de s1 se "MOVIÓ" a s2

// println!("{}", s1); // ¡ESTO DARÍA ERROR! s1 ya no es dueño del dato.
println!("s2 dice: {}", s1); // Esto funciona perfectamente

// Si queremos usarlo sin perder la propiedad, usamos referencias (&)
fn calcular_longitud(s: &String) -> usize {
    s.len()
} // Aquí 's' sale de scope, pero como es una referencia, no se borra el dato original

let len = calcular_longitud(&s2);
println!("La longitud de '{}' es {}.", s2, len);
```

6. Estructuras (Structs)

Las structs te permiten crear tipos de datos personalizados agrupando valores relacionados. Son similares a los objetos en otros lenguajes, pero sin herencia.

Existen tres tipos de structs en Rust:

    • Classic Structs: Con nombres de campos (las más comunes).
    • Tuple Structs: Básicamente tuplas con nombre.
    • Unit-like Structs: Sin campos, útiles para implementar rasgos (traits) en un tipo.

```rust
// Definición de una Classic Struct
struct Usuario {
    nombre: String,
    email: String,
    activo: bool,
    cuenta_clicks: u64,
}

// Instanciación
let mut usuario1 = Usuario {
    email: String::from("alguien@ejemplo.com"),
    nombre: String::from("Romar"),
    activo: true,
    cuenta_clicks: 1,
};

// Acceso y modificación (requiere que la instancia sea 'mut')
usuario1.email = String::from("nuevo_email@ejemplo.com");

println!("Usuario: {} <{}>", usuario1.nombre, usuario1.email);
```

7. Enumeraciones (Enums) y Pattern Matching

Los enums en Rust son mucho más potentes que en otros lenguajes. No solo enumeran variantes, sino que cada variante puede almacenar datos.

Para manejar un Enum, usamos la expresión match. Rust te obliga a cubrir todas las posibilidades (exhaustividad), lo que elimina los errores de "casos no controlados".

```rust
enum Mensaje {
    Salir,
    Mover { x: i32, y: i32 }, // Variante con struct anónima
    Escribir(String),          // Variante con un String
    CambiarColor(i32, i32, i32), // Variante tipo tupla
}

fn procesar_mensaje(msg: Mensaje) {
    match msg {
        Mensaje::Salir => println!("Cerrando aplicación..."),
        Mensaje::Escribir(texto) => println!("Mensaje recibido: {}", texto),
        Mensaje::Mover { x, y } => println!("Mover a posición X:{}, Y:{}", x, y),
        _ => println!("Otro tipo de mensaje recibido"), // El guion bajo es el caso por defecto
    }
}

let m = Mensaje::Escribir(String::from("Hola Rust"));
procesar_mensaje(m);
```

8. El Enum más importante: Option<T>

En Rust no existe el valor null. Para representar la ausencia de un valor, usamos el Enum estándar Option. Esto evita los famosos errores de "puntero nulo".

• Some(T): Hay un valor de tipo T.
• None: No hay valor.

```rust
fn dividir(dividendo: f64, divisor: f64) -> Option<f64> {
    if divisor == 0.0 {
        None // En lugar de lanzar un error, devolvemos "Nada"
    } else {
        Some(dividendo / divisor)
    }
}

let resultado = dividir(10.0, 2.0);

match resultado {
    Some(valor) => println!("Resultado: {}", valor),
    None => println!("¡Error: División por cero!"),
}
```

9. Implementación de Métodos (impl)

Puedes definir comportamiento para tus structs y enums usando bloques impl. El primer parámetro suele ser &self, que representa la instancia de la estructura.

```rust
struct Rectangulo {
    ancho: u32,
    alto: u32,
}

impl Rectangulo {
    // Esto es un método
    fn area(&self) -> u32 {
        self.ancho * self.alto
    }

    // Esto es una función asociada (similar a un método estático)
    fn cuadrado(medida: u32) -> Rectangulo {
        Rectangulo { ancho: medida, alto: medida }
    }
}

let rect = Rectangulo::cuadrado(10);
println!("El área del cuadrado es: {} px2", rect.area());
```

10. Gestión de Errores con Result

En Rust, los errores no son "excepciones" que interrumpen el programa de forma invisible. Los errores son valores que el compilador te obliga a manejar.

El Enum Result<T, E> tiene dos variantes:

    • Ok(T): La operación tuvo éxito y contiene el valor T.
    • Err(E): La operación falló y contiene el error E.

    El operador ?: Es una de las herramientas más potentes de Rust. Si una operación devuelve Err, el operador ? hace que la función actual retorne ese error inmediatamente, propagándolo hacia arriba.

```rust
#[derive(Debug)]
enum ErrorCajero {
    FondosInsuficientes,
    PinInvalido,
}

fn retirar_dinero(cantidad: u32, saldo: u32) -> Result<u32, ErrorCajero> {
    if cantidad > saldo {
        return Err(ErrorCajero::FondosInsuficientes);
    }
    Ok(saldo - cantidad)
}

// Ejemplo de uso con Match
let mi_saldo = 100;
let intento_retiro = retirar_dinero(150, mi_saldo);

match intento_retiro {
    Ok(nuevo_saldo) => println!("Retiro exitoso. Saldo restante: {}", nuevo_saldo),
    Err(e) => println!("Error al retirar: {:?}", e),
}
```

11. Organización del Código: Módulos y Crates

A medida que tu proyecto crece, no puedes tener todo en un solo archivo. Rust utiliza un sistema de módulos para controlar el alcance (scope) y la privacidad.

    • Crate: Es la unidad de compilación más pequeña (un archivo .rs o una librería completa).
    • Module (mod): Permite agrupar funciones, structs y enums dentro de un crate.
    • Privacidad: En Rust, todo es privado por defecto. Debes usar la palabra clave pub para hacer que algo sea accesible desde fuera.

```rust
// Definición de un módulo interno
mod comunicacion {
    // Este módulo es público para que otros puedan entrar
    pub mod mensajeria {
        // Esta función debe ser 'pub' para ser llamada desde fuera
        pub fn enviar_alerta(mensaje: &str) {
            println!("ALERTA DE SISTEMA: {}", mensaje);
        }
        
        // Esta función NO es pública, solo se ve dentro de 'mensajeria'
        fn log_interno() {
            println!("Log guardado.");
        }
    }
}

// Uso de rutas (paths) para llamar a la función
use comunicacion::mensajeria::enviar_alerta;

fn main() {
    enviar_alerta("Fuga de memoria detectada");
    
    // También se puede usar la ruta completa
    comunicacion::mensajeria::enviar_alerta("Reintento de conexión");
}

main(); // Ejecución manual para el notebook
```

Resumen de Arquitectura

Para recapitular cómo se conecta todo en una aplicación real:

    • Cargo.toml: Define las dependencias (Crates externas).
    • src/main.rs: Punto de entrada de la aplicación.
    • Módulos: Separan la lógica de negocio (ej. mod base_datos, mod auth).
    • Enums/Structs: Definen la forma de los datos.
    • Result/Option: Aseguran que cada posible fallo tenga una respuesta controlada.

12. Colecciones Dinámicas

A diferencia de los arrays (que tienen un tamaño fijo), las colecciones en Rust viven en el Heap (montículo) y pueden crecer o encogerse durante la ejecución.

Vectores (Vec<T>)
Es la colección más común. Imagínala como un array redimensionable.

    • Se crea con Vec::new() o la macro vec![].
    • Solo puede almacenar elementos del mismo tipo.

HashMaps (HashMap<K, V>)
Almacena pares de clave-valor. Es ideal para búsquedas rápidas mediante un identificador único.

```rust
use std::collections::HashMap;

// --- Trabajando con Vectores ---
let mut puntuaciones = vec![10, 25, 30];
puntuaciones.push(45); // Añadir elemento

println!("Segunda puntuación: {}", puntuaciones[1]);

// --- Trabajando con HashMaps ---
let mut inventario = HashMap::new();

inventario.insert(String::from("Manzanas"), 10);
inventario.insert(String::from("Bananas"), 5);

// Acceso seguro usando Option (visto anteriormente)
let nombre_fruta = String::from("Manzanas");
match inventario.get(&nombre_fruta) {
    Some(cantidad) => println!("Hay {} {}", cantidad, nombre_fruta),
    None => println!("No hay stock de {}", nombre_fruta),
}
```

13. Closures e Iteradores

Los Closures son funciones anónimas que puedes guardar en una variable o pasar como argumento a otras funciones. Su característica principal es que pueden "capturar" variables de su entorno.

Los Iteradores permiten procesar secuencias de elementos de forma eficiente (y a menudo más rápida que un bucle for manual).

```rust
let factor = 2;

// Definición de un Closure que captura la variable 'factor'
let multiplicar = |n: i32| n * factor;

let numero = 5;
println!("{} multiplicado por {} es: {}", numero, factor, multiplicar(numero));

// --- Uso de Iteradores ---
let numeros = vec![1, 2, 3, 4, 5];

// Programación funcional: filtrar y transformar
let suma_pares_duplicados: i32 = numeros.iter()
    .filter(|&&x| x % 2 == 0) // Filtramos solo pares
    .map(|x| x * 2)           // Los duplicamos
    .sum();                   // Los sumamos todos

println!("Resultado del procesamiento: {}", suma_pares_duplicados);
```
